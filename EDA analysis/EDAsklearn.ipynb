## How to transform numpy array to pandas data frame

```python
import pandas as pd 
import seaborn as sb 
import matplotlib.pyplot as plt 

from sklearn.datasets import fetch_california_housing
```

## this fetched dataset of california housing is in numpy array form , for better handling we convert it to pandas dataframe 

```python
data = fetch_california_housing()
print(type(data))
print(data)
```

# load the dataset 

```python
housing = fetch_california_housing(as_frame=True)
print(type(housing))
data = housing.frame
print(type(data))
print(data)
```

# display the first few row of the dataset

```python
print(data.head())
```

# get the shape of the dataset(no of rows and coloumns)

```python
print(data.shape)
```

# calculate summary statistics (mean,median,mod) for the numerical coloumns of the dataset

```python
print(data.mean())
```

```python
print(data.median())
```

```python
print(data.std())
```

```python
print(data.describe())
```

# distribution of key features 

```python
keyfeatures = ['MedInc','HouseAge','MedHouseVal']
for features in keyfeatures:
    plt.hist(data[features])
    plt.grid(True)
    plt.xlabel('Features')
    plt.show()

```

```python
keyfeatures=['MedInc','HouseAge','MedHouseVal']
for features in keyfeatures:
    plt.plot(data[features])
    plt.grid(True)
    plt.show()
```

# Display the target variable 'MedHouseVal'

```python
plt.hist(data['MedHouseVal'])
plt.xlabel('Distribution of target variable')
plt.grid(True)
plt.show()
```

```python
sb.histplot(data['MedHouseVal'],kde=True)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sb.histplot(data['MedHouseVal'], kde=True, bins=30, color='purple')
plt.title('Distribution of Median House Value (Target)')
plt.xlabel('MedHouseVal')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Pairplot to visualize pairwise relationships

```python
sb.pairplot(data=data[keyfeatures],hue='MedHouseVal')
sb.set_style('whitegrid')
plt.show()
```
