```python
sb.scatterplot(data = data,x='petal.length',y='sepal.length',hue='variety')
plt.show()
```

```python
sb.set_style('whitegrid')
sb.scatterplot(data = data,x = 'petal.length',y = 'petal.width',hue = 'variety')
plt.show()
```

```python
sb.set_style('whitegrid')
sb.pairplot(data=data,hue='variety')
plt.show()
```

```python
sb.boxplot(data = data,x='variety',y='sepal.width')
sb.set_style('whitegrid')
plt.show()
```

```python
sb.violinplot(data=data,x='variety',y='sepal.width')
plt.show()
```
